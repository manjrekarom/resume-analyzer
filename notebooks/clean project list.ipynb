## First Projects List

```python
import os
```

```python
with open('../datasets/projects.list', 'r') as f:
    txt = f.read()
```

```python
result_lines = []
for line in txt.split('\n'):
    idx = line.find(':')
    if idx != -1:
        result_lines.append(line[idx+1:])
```

```python
with open('../datasets/projects.list', 'w+') as f:
    f.write('\n'.join(result_lines))
```

## Second Projects List

```python
with open('../datasets/projects2.list', 'r') as f:
    txt2 = f.read()
```

```python
result_lines = []
for line in txt2.split('\n'):
    if not line.strip() or line.endswith(':') or line.startswith('Part'):
        continue
    result_lines.append(line)
```

```python
result_lines
```

```python
with open('../datasets/projects_final.list', 'a+') as f:
    f.write('\n'.join(result_lines))
```
